```python
#라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#데이터셋 불러오기
df = pd.read_csv('International football results from 1872 to 2023/results.csv')
```

```python
#경기 날짜와 골수와 합차 구하기
date_array = df['date']
goals_array = df['home_score'] + df['away_score']
goal_difference_array = abs(df['home_score'] - df['away_score'])
```

```python
#꺾은선 그래프 그리기
plt.plot(date_array, goals_array, label='Goals', color='green', marker='o')
plt.plot(date_array, goal_difference_array, label='Goal Difference', color='orange', linestyle='--', marker='x')

plt.title('Goals and Goal Difference Over Time')
plt.xlabel('Date')
plt.ylabel('Goals')
plt.legend()

plt.figure(figsize=(16, 8))
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```
